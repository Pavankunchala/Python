```python
import seaborn as sns
import matplotlib.pyplot as plt
sns,set()
tips = sns.load_dataset("tips")
sns.relplot(x = 'total_bill', y = 'tip',hue = 'smoker',col = 'time',size = 'day',style = 'size',data = tips)
```

```python
fmri   =sns.load_dataset('fmri')
sns.relplot(x = 'timepoint' , y= 'signal', col = 'region', kind='line',hue='event', style='event', data=fmri)
```
